# Building a diffraction grating

* Step one: build an interference pattern for 3 sources
* Step two: build an interference pattern for 4 sources
* Step three: build an interference pattern for n sources
* Step four: scale by single wide slit difraction pattern.

```python
# Define the wavefunction...
#import numpy as np
from mpl_toolkits.mplot3d import axes3d
#from matplotlib import cm

size_of_font=15

n_slits = 3 #Number of slits
a = 0.2 # Width of slit
b = 10 # Separations between slit
l_x = 12 # Distance across screen, x direction
l_y = 600 # how high do i want the diffraction pattern to appear

x = np.linspace(-l_x,l_x,2000)
y = np.linspace(0,l_y,10)

L = 50 #Distance to screen
wavelength_r = 1# Wavelength of radiation
wavelength_g = 1
wavelength_b = 1
omega_t = np.pi # Alpha parameter for light.
A_0 = 1 # Intensity of each source
sources =  10

phase_offset_r = (2*np.pi*b*x)/(wavelength_r*L)
phase_offset_g = (2*np.pi*b*x)/(wavelength_g*L)
phase_offset_b = (2*np.pi*b*x)/(wavelength_b*L)

def amp_x(delta, sources):
    A_i = 0
    for i in np.arange(sources):
        A_i = A_i+np.cos(omega_t+ (i*delta))
    return A_i

def amp_y(delta, sources):
    A_j = 0
    for i in np.arange(sources):
        A_j = A_j+np.cos(omega_t+ (i*delta))
    return A_j

def mod_A(delta,sources):
    A = ((amp_y(delta,sources))**2 + (amp_x(delta,sources))**2)**1
    return A

def doubleslit(x,y):
    z = (((np.sinc(2*np.pi*a*x))**2)*(np.cos(2*np.pi*b*x))**2)**0.5
    return z

def singleslit(x,y):
    z = (((np.sinc(2*np.pi*a*x))**2))**0.5
    return z



X,Y = np.meshgrid(x,y) # Links the X and Y data points together
z = np.array([doubleslit(x,y) for x,y in zip(np.ravel(X),np.ravel(Y))]) # Calculates the z value for each X/Y combination
Z = z.reshape(X.shape) # Reshapes the Z data array into the same shape as X

DELTA,Y = np.meshgrid(phase_offset_r,y)

z2 = np.array([mod_A(phase_offset_r,2) for phase_offset_r,y in zip(np.ravel(DELTA),np.ravel(Y))])
Z2 = z2.reshape(X.shape)

z3 = np.array([mod_A(phase_offset_r,3) for phase_offset_r,y in zip(np.ravel(DELTA),np.ravel(Y))])
Z3 = z3.reshape(X.shape)

z10 = np.array([mod_A(phase_offset_r,10) for phase_offset_r,y in zip(np.ravel(DELTA),np.ravel(Y))])
Z10 = z10.reshape(X.shape)

#X,Y,Z = ([x],[y],[z])


#fig = plt.figure(figsize=(7,5))
figure, axis = plt.subplots(3, 1,figsize=(10,10))




#plt.contourf(X,Y,Z,100,cmap='Blues',alpha=0.5)
#plt.plot(x,mod_A(phase_offset_r,sources),color='red')
#plt.plot(x,mod_A(phase_offset_g,sources),color='green')
#plt.plot(x,mod_A(phase_offset_b,sources),color='blue')

axis[0].plot(x,mod_A(phase_offset_r,2),color='blue')
axis[0].set_yticklabels([])
axis[0].yaxis.set_ticks_position('none') 
axis[0].set_xlabel('Position on screen')
axis[0].set_ylabel('Intensity')
axis[0].set_title('Two slit diffraction pattern')
axis[0].contourf(X,Y,Z2,100,cmap='Blues',alpha=0.5)
axis[0].set_ylim(0,1.3*np.max(mod_A(phase_offset_r,2)))



axis[1].plot(x,mod_A(phase_offset_r,3),color='blue')
axis[1].set_yticklabels([])
axis[1].yaxis.set_ticks_position('none') 
axis[1].set_xlabel('Position on screen')
axis[1].set_ylabel('Intensity')
axis[1].set_title('Three slit diffraction pattern')
axis[1].contourf(X,Y,Z3,100,cmap='Blues',alpha=0.5)
axis[1].set_ylim(0,1.3*np.max(mod_A(phase_offset_r,3)))




axis[2].plot(x,mod_A(phase_offset_r,10),color='blue')
axis[2].set_yticklabels([])
axis[2].yaxis.set_ticks_position('none') 
axis[2].set_xlabel('Position on screen')
axis[2].set_ylabel('Intensity')
axis[2].set_title('Ten slit diffraction pattern')
axis[2].contourf(X,Y,Z10,100,cmap='Blues',alpha=0.5)
axis[2].set_ylim(0,1.3*np.max(mod_A(phase_offset_r,10)))








#plt.plot(x,singleslit(x,y)**2,color='blue',linestyle='dotted')



#plt.xlabel('$X$ ',fontsize=size_of_font)
#plt.ylabel('Relative intensity',fontsize=size_of_font)
#plt.xticks([0],('0'))

plt.tight_layout()
#plt.savefig("ch13-diffraction-patterns1.svg",transparent=True)
#plt.savefig("ch13-diffraction-patterns1.pdf",transparent=True)
plt.show()
```

# Radial diffraction from optical apertures
Basically taking sinc function and turning into a radial distribution function

```python
# Specify x and y to start with
number_of_points = 400
value_range = 1.5
x_coord = np.linspace(-value_range,value_range,number_of_points)
y_coord = np.linspace(-value_range,value_range,number_of_points)

X,Y = np.meshgrid(x_coord,y_coord)

x_offset = 0
y_offset = 0

def sinc_func(x,y):
    r = np.sqrt((x-x_offset)**2 + (y-y_offset)**2)
    func = np.sinc(2*r)
    return func

def radius(x,y):
    radius = np.sqrt((x)**2 + (y)**2)
    return radius

z_int = np.array( [sinc_func(x,y) for x,y in zip(np.ravel(X),np.ravel(Y))])
ZINT = z_int.reshape(X.shape)


plt.figure(figsize=(10,10))
plt.axes().set_aspect('equal')#, 'datalim') #equalises the aspect ratio

levels = np.linspace(0,0.4,100)

cs = plt.contourf(X,Y,(ZINT**2)**0.5,levels=levels, cmap='gray')
#plt.yticks([0],('0'))
#plt.xticks([0],('0'))
plt.axhline(y=0,color='red',linestyle='dotted',alpha=1)

#plt.colorbar()

#plt.axvline(x=0,color='white',linestyle='dotted',alpha=0.5)
#plt.colorbar();
#savefig('aw.pdf') # Needs to go above show() otherwise we'll get a blank PDF


#plt.plot(radius(x_coord,y_coord),sinc_func(x_coord,y_coord))

for c in cs.collections:
    c.set_rasterized(True)

plt.tight_layout()
#plt.savefig("ch13-singlesource-diffraction.svg",transparent=True)
#plt.savefig("ch13-singlesource-diffraction.pdf",transparent=True)
plt.show()
```

```python
# Specify x and y to start with
number_of_points = 400
value_range = 1.5
x_coord = np.linspace(-value_range,value_range,number_of_points)
y_coord = np.linspace(-value_range,value_range,number_of_points)

X,Y = np.meshgrid(x_coord,y_coord)

offset = 0.35

x_offset1 = -offset
y_offset1 = -0

x_offset2 = offset
y_offset2 = 0

def sinc_func1(x,y):
    r = np.sqrt((x-x_offset1)**2 + (y-y_offset1)**2)
    func = np.sinc(2*r)
    return func

def sinc_func2(x,y):
    r = np.sqrt((x-x_offset2)**2 + (y-y_offset2)**2)
    func = np.sinc(2*r)
    return func

def radius(x,y):
    radius = np.sqrt((x)**2 + (y)**2)
    return radius

z_int1 = np.array( [sinc_func1(x,y) for x,y in zip(np.ravel(X),np.ravel(Y))])
ZINT1 = z_int1.reshape(X.shape)

z_int2 = np.array( [sinc_func2(x,y) for x,y in zip(np.ravel(X),np.ravel(Y))])
ZINT2 = z_int2.reshape(X.shape)

z_tot = ZINT1+ZINT2

plt.figure(figsize=(10,10))
plt.axes().set_aspect('equal')#, 'datalim') #equalises the aspect ratio
levels = np.linspace(0,0.8,100)

cs = plt.contourf(X,Y,(z_tot**2)**0.5,levels=levels, cmap='gray')
#plt.yticks([0],('0'))
#plt.xticks([0],('0'))
plt.axhline(y=0,color='red',linestyle='dotted')
#plt.axvline(x=0,color='blue',linestyle='dotted')
#plt.colorbar();
#savefig('aw.pdf') # Needs to go above show() otherwise we'll get a blank PDF

for c in cs.collections:
    c.set_rasterized(True)

plt.tight_layout()
plt.savefig("ch13-opticalresolutionlimit3.svg",transparent=True)
plt.savefig("ch13-opticalresolutionlimit3.pdf",transparent=True)
#plt.plot(radius(x_coord,y_coord),sinc_func(x_coord,y_coord))
plt.show()
```

```python
# Specify x and y to start with
number_of_points = 400
value_range = 4
scalar = 1


x_coord = np.linspace(-value_range,value_range,number_of_points)
y_coord = np.linspace(-1*value_range,1*value_range,number_of_points)

X,Y = np.meshgrid(x_coord,y_coord)

offset = 1

x_offset1 = -offset
y_offset1 = -0

x_offset2 = offset
y_offset2 = 0


def sine1(x,y):
    r = np.sqrt((x-x_offset1)**2 + (y-y_offset1)**2)
    func = np.sin(2*np.pi*r*scalar)
    return func

def sine2(x,y):
    r = np.sqrt((x-x_offset2)**2 + (y-y_offset2)**2)
    func = np.sin(2*np.pi*r*scalar)
    return func

def sine3(x,y):
    r = np.sqrt((x-x_offset2)**2 + (y-y_offset2)**2)
    func = np.sin(2*np.pi*r*scalar + np.pi/1)
    return func

def radius(x,y):
    radius = np.sqrt((x)**2 + (y)**2)
    return radius

z_int1 = np.array( [sine1(x,y) for x,y in zip(np.ravel(X),np.ravel(Y))])
ZINT1 = z_int1.reshape(X.shape)

z_int2 = np.array( [sine3(x,y) for x,y in zip(np.ravel(X),np.ravel(Y))])
ZINT2 = z_int2.reshape(X.shape)

z_tot = ZINT1+ZINT2

plt.figure(figsize=(10,10))
plt.axes().set_aspect('equal')#, 'datalim') #equalises the aspect ratio
#levels = np.linspace(0,0.8,100)
#plt.xlim(-4,4)
#plt.ylim(-4,4)


#plt.contourf(X,Y,(ZINT1)**1,100, cmap='gray',alpha=1)
#plt.contourf(X,Y,(ZINT2)**1,100, cmap='gray',alpha=1)
cs=plt.contourf(X,Y,(z_tot)**2,100, cmap='gray',alpha=1)

# Plot the maxima
circles=np.arange(0,10,1)
theta = np.linspace(0,2*np.pi,200)

for circle in circles:
    radius = circle+0.25
    x_circ = radius*np.cos(theta)-offset
    y_circ = radius*np.sin(theta)
    plt.plot(x_circ,y_circ,color='blue')

for circle in circles:
    radius = circle
    x_circ = radius*np.cos(theta)+offset
    y_circ = radius*np.sin(theta)
    plt.plot(x_circ,y_circ,color='orange')
    
for circle in circles:
    radius = circle-0.25
    x_circ = radius*np.cos(theta)-offset
    y_circ = radius*np.sin(theta)
    plt.plot(x_circ,y_circ,color='blue',linestyle='dotted')

for circle in circles:
    radius = circle-0.5
    x_circ = radius*np.cos(theta)+offset
    y_circ = radius*np.sin(theta)
    plt.plot(x_circ,y_circ,color='orange',linestyle='dotted')

plt.plot(0.87,1.24,'ro',markersize='10')

#plt.yticks([0],('0'))
#plt.xticks([0],('0'))
#plt.axhline(y=0,color='red',linestyle='dotted')

#plt.axvline(x=0,color='blue',linestyle='dotted')
#plt.colorbar();
#savefig('aw.pdf') # Needs to go above show() otherwise we'll get a blank PDF

for c in cs.collections:
    c.set_rasterized(True)

plt.xlim(-4,4)
plt.ylim(-4,4)



plt.tight_layout()
#plt.savefig("ch13-opticalresolutionlimit3.svg",transparent=True)
plt.savefig("interference1.png",transparent=True)
##plt.plot(radius(x_coord,y_coord),sinc_func(x_coord,y_coord))
plt.show()
```
